```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Sample data for token lengths and accuracies
x_axis = np.array([0, 8,
                   16, 32,
                   64, 128,
                   256])
y_axis = np.array([0.1, 0.5,
    0.3, 0.4,
    0.8, 0.9,
    0.9])
error = np.array([0.05, 0.08,
                  0.02, 0.03,
                  0.03, 0.02,
                  0.03])

plt.figure(figsize=(6, 4))
plt.plot(x_axis, y_axis, 'o-', color='brown', linewidth=1.5, markersize=6)
plt.fill_between(x_axis, y_axis - error, y_axis + error, color='brown', alpha=0.2)

# Set labels and title
plt.xlabel("Token Length", fontsize=15)#, weight='bold')
plt.ylabel("CLIP Image Similarity", fontsize=15)#, weight='bold')
plt.ylim(0, 1.1)

# Aesthetics for an academic-style plot
plt.grid(True, which='both', linestyle='--', linewidth=0.5, alpha=0.7)
plt.xticks(fontsize=13)
plt.yticks(fontsize=13)
sns.despine()  # Removes the top and right border for a cleaner look

# Save the figure as a high-quality image for publication
plt.tight_layout()
plt.savefig("academic_style_plot.pdf", dpi=300)
plt.show()
```
